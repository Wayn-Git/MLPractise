```python
import pandas as pd
import numpy as np

def outlier_report(df, whisker=1.5):
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    report = []
    
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - whisker * IQR
        upper = Q3 + whisker * IQR
        
        n_outliers = df[(df[col] < lower) | (df[col] > upper)].shape[0]
        pct_outliers = n_outliers / len(df) * 100
        
        if n_outliers > 0:  # only show columns that actually have outliers
            report.append({
                'column': col,
                'outliers': n_outliers,
                '%': round(pct_outliers, 2),
                'max_value': df[col].max(),
                'upper_bound': round(upper, 2)
            })
    
    return pd.DataFrame(report).sort_values('outliers', ascending=False)

# RUN THIS → instant full report
outlier_report(df).reset_index(drop=True)
```
